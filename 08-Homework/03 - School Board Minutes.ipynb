````python
import requests
from bs4 import BeautifulSoup
````

````python
url = 'http://www.mineral.k12.nv.us/pages/School_Board_Minutes'
response = requests.get(url, verify=False)
school = BeautifulSoup(response.text, 'html.parser')
````

````python
school.head()
````

````python
import pandas as pd
minutes = school.find_all('p')

min_list = []

for minute in minutes:
    links = minute.find_all('a')
    for link in links:
        min_dict = {}
        min_dict['date'] = link.text
        min_dict['url'] = link['href']
        min_list.append(min_dict)

df = pd.DataFrame(min_list)

df['date'] = pd.to_datetime(df['date'], errors='coerce')

df = df.dropna(subset=['date'])

df
````
