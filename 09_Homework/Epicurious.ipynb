```python
rows = []
for page_num in range(1, 11):
    url = f"https://www.epicurious.com/search/cucumbers?page={page_num}"
#    print("Now scraping", url)
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    results = doc.find_all(class_="results-group")[0].find_all('article')
    for result in results:
        row = {}
        try:
            row['tag'] = result.find(class_="tag").text.strip()
        except:
            pass
        try:
            row['title'] = result.find(class_="hed").text.strip()
        except:
            pass
        try:
            row['summary'] = result.find(class_="dek").text.strip()
        except:
            pass
        try:
            row['rating'] = result.find(itemprop="ratingValue").text.strip()
        except:
            pass
        try:
            row['make_again'] = result.find(class_="make-again-percentage").text.strip()
        except:
            pass
        try:
            row['url'] = result.find('a')['href']
        except:
            pass
        rows.append(row)
        print(row['url'])
print("Building our dataframe")
df = pd.DataFrame(rows)
df.head()
```

```python
def scrape_page(row):
    if row['tag'] == 'recipe':
        url = f"https://www.epicurious.com{row['url']}"
        print("Scraping", url)
        response = requests.get(url)
        doc = BeautifulSoup(response.text)
        pages = []
        details = doc.find_all(class_="main-content")
        for detail in details:
            page = {}
 #           try:
 #               page['title'] = detail.find_all(class_="ingredients").text.strip()
 #           except:
 #               pass
            try:
                page['ingredients'] = detail.find_all(class_="ingredients").text.strip()
            except:
                pass
            try:
                page['direction'] = detail.find(class_="preparation-steps").text.strip()
            except:
                pass
            try:
                page['tags'] = detail.find(class_="tags").text.strip()
            except:
                pass
            pages.append(page)
            return pd.Series(page)
```

```python
scraped_df = df.apply(scrape_page, axis=1)
scraped_df
```

```python
df.merge(scraped_df, left_index=True, right_index=True)

```
