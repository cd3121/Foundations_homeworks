```python
rows = []
for page_num in range(1, 5):
    url = f"http://www.metrolyrics.com/rem-alpage-{page_num}.html"
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    songs = doc.find('tbody').find_all('tr')
    for song in songs:
        row ={}
        row['title'] = song.find(class_="title").text.strip()
        row['url'] = song.find('a')['href']
        row['year'] = song.find_all('td')[2].text.strip()
        row['popularity'] = song.find(class_='bar').find('span')['style']
        rows.append(row)

print("Building our dataframe")
df = pd.DataFrame(rows)
df
df.shape
```

```python
df.to_csv("metrolyrics.csv", index=False)
```

```python
def get_lirics(url):    
    response = requests.get(url)
    doc = BeautifulSoup(response.content, 'html.parser')
    try:
        lyrics = doc.find_all('p', attrs={"class":'verse'})
        lirics = '\n'.join([line.text for line in lyrics])
        return lirics
    except:
        pass
```

```python
df["Lirics"] = df.url.apply(get_lirics)
```
